> Q5. Using a while loop, verify if the number A is purely divisible by
> number B and if so then how many times it can be divisible.
>
> Ans - def calculate_divisibility_count(A, B):
>
> count = 0
>
> while A % B == 0:
>
> count += 1
>
> A = A // B \# Integer division to reduce A
>
> return count
>
> \# Example values
>
> number_A = 64
>
> number_B = 2
>
> divisibility_count = calculate_divisibility_count(number_A, number_B)
>
> if divisibility_count \> 0:
>
> print(f"{number_A} is divisible by {number_B} and can be divided
> {divisibility_count} times.")
>
> else:
>
> print(f"{number_A} is not divisible by {number_B}.")
