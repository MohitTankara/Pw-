> Q11. What is the formula for calculating the sample mean? Provide an
> example calculation for a dataset.
>
> Ans -  
> **The formula for calculating the sample mean is:**
>
> **Sample Mean (x̄ ) = Σx / n**
>
> **where:**  
> **x̄ = Sample mean**  
> **Σx = Sum of all values in the dataset**  
> **n = Number of values in the sample**
>
> **Example calculation:**  
> **Let's calculate the sample mean for the following dataset:**  
> **\[12, 15, 18, 20, 22, 25\]**
>
> **Step 1: Add up all the values in the dataset.**
>
> **Sum of the values (Σx) = 12 + 15 + 18 + 20 + 22 + 25 = 112**
>
> **Step 2: Count the number of values in the dataset (n).**  
> **Number of values (n) = 6**
>
> **Step 3: Apply the formula for the sample mean.**
>
> **Sample Mean (x̄ ) = Σx / n = 112 / 6 = 18.67**
>
> **The sample mean for the given dataset is approximately 18.67.**
